### Importing the ABC method from abc module + abstractmethod

#### Abstractmethods are used in implementation of the abstract class. It can not be assigned as the object and it can not be used by other developres. 

#### If abstractclass has abstractmethods, they must be implemented in the children classes as well. They do not need to be defined under abstractclass

```python
# class vehicle defined by me 
# other define types

class Vehicle(ABC):
    def __init__(self,car_name,id_num):
        self.car_name = car_name
        self.id_num = id_num

    @abstractmethod
    def car_call(self):
        return self.__doc__
```

How should I inherit this abstract class -> concrete class?

#### Step-by-Step Implementation

- Inherit from the Abstract Class: Define your concrete class and pass the abstract class name into its definition. -> Concrete

- Override the Abstract Method: Define the method using the exact name specified in the parent class. 

Here abstractclass method is called carr_call

- Write the Logic: **Omit the @abstractmethod decorator** in the concrete class and write your actual code instead of using pass.

```python

class Concrete(Vehicle):
    def __init__(self,model):
        self.model = self
    def car_call(self):
        return super().car_call()
```

### Important things to remember:


#### 1) Abstract classes in Python are allowed to have normal, fully implemented methods and standard class/instance attributes. Subclasses inherit these automatically without needing to redefine them

#### 2) My main abstractclass can be inherited by the downstream abstractclass, and next concrete class will inherit directly from this last abstractclass.

```python
# checking the inheritance of the 2 downstream abstractclasses

class Person(ABC):
    def __init__(self,name,age,salary):
        self.name = name
        self.age = age
        self.salary = salary
    def __str__(self):
        return f"Person name is {self.name}, aged {self.age} with salary {self.salary}"
    @abstractmethod
    def greet(self):
        print(f"hello person {self.name}")


# Second abstract class

class Worker(Person):
    def __init__(self,name,age,id,salary):
        self.salary= salary
        self.id = id
        self.age = age
        self.name = name
        super().__init__(name,age,salary)
    def greet_this(self):
        print(f"Hello this person {self.name},{self.age} with id {self.id}")

# First concrete class

class Human(Worker):
    def __init__(self,hair,name,age,id,salary):
        self.salary = salary
        self.hair= hair
        super().__init__(name,age,id,salary)
    def greet_this(self):
        return super().greet_this()
    def greet(self):
        return super().greet()
```

```python
petra = Human('brown',"Petra",25,1921,10000)
petra.hair
petra.greet_this()
petra.greet()
print(petra)
```

#### Description

##### You need to create an abstract class Vehicle. Classes Car, Motorcycle, Truck, and Bus are inherited from Vehicle and already implemented.

Class Vehicle accepts the following parameters:

    brand_name -> str (e.g. Honda)
    year_of_issue -> int (e.g. 2020)
    base_price -> int (e.g. 1_000_000)
    mileage -> int (e.g. 10_000)

The following methods should be implemented:

    vehicle_type - returns str - a type of the vehicle in the following pattern brand_name + name of the class. For example: Toyota Car, Suzuki Motorcycle;
    is_motorcycle returning a boolean value depends on the number of wheels (2 wheels -> motorcycle, so the method should return True);
    purchase_price - returns the price of the vehicle: (base_price - 0.1 * mileage). If the resulting price is less than 100_000, the method should return 100_000.

Put the following decorators where necessary and if it is necessary:

abstractmethod, classmethod, staticmethod, property, and other decorators.

## Intuition:

- abstract class vehicles implement input parameters and methods

- methods that must be same in all child classes will be @abstractmethods

- methods that must be defined in each class separately will be standard methods or property

### **1) vehicle_type -> this method is uniform in each child class, so it will be concrete**

### **2) is_motorcycle -> this will be the property defined within each class, so it will be abstract**

### **3) purschase_price -> for each class, must be abstract**

```python
from abc import ABC,abstractmethod
```

```python
class Vehicle(ABC):
    def __init__(self,brand_name,year_of_issue,base_price,mileage):
        self.brand_name = brand_name
        self.year_of_issue = year_of_issue
        self.base_price = base_price
        self.mileage = mileage

    def vehicle_type(self):
        return f"{self.brand_name}, {self.__class__.__name__}"
    

    @property
    @abstractmethod
    def wheels(self):
        pass
    
    def is_motorcycle(self):
        return self.wheels == 2
    def purscase_price(self):
        price = self.base_price - (0.1*self.mileage)

        return max(10000,price)
```

```python

class Car(Vehicle):
    def __init__(self,brand_name,year_of_issue,base_price,mileage):
        super().__init__(brand_name,year_of_issue,base_price,mileage)
    @property
    def wheels(self):
        return 4
```

```python
car = Car('honda',1998,19999,100)
```

```python
print(
            f"Vehicle type={car.vehicle_type()}\n"
            f"Is motorcycle={car.is_motorcycle()}\n"
            f"Purchase price={car.purscase_price()}\n"
        )
```
