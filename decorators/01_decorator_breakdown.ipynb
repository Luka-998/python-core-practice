### **Decorators**

Decorators are function that are used to modify the main function behaviour such as making additional modifications before/after 

main function execution, without changing the function original code

__Simple example of the decorator:__

```python
# first we define a decorator structure: 
# decorator function accepts our function that we are about to modify

def decorate_this(func):
    def wrapper(*args): # the wrapper function wraps the original function that we are calling
        """ docstrig wrom the wrapper"""
        print("Starting....")
        result = func(*args)
        print(result)
        print("End.")
        return result

    return wrapper
```

### Decorator call 

```python
@decorate_this
def double_this(x):
    """My main function docs"""
    return x**2
d = double_this(4)
```

#### Execution path:

##### 1) **double_this** (main function) is called with argument x: d = double_this(4)

##### 2) Then @decorate_this does the following: __double_this = decorate_this(double this)__. Here main function __double_this__ becomes decorator argument (func)!

##### 3) It goes to inner function **wrapper** which takes any amount of positional *args and call that function inside with additional modifications. It does not execute the function here. It just makes a function that will be executed later.


##### 4) **wrapper** return result, and **decorate_this** returns the wrapper. In that terms my original function **double_this** becomes  = **wrapper**

#### From here:

- double_this function points at wrapper. Wrapper remembers the original function via **Closure**

- With closure principle, decorator remember the orignal function and allows it execution

- This also mean that if we try to print out original function name and doc, we will get wrapper name and doc! Lets see:

```python
print(double_this.__name__) # we expect original function name but we will get decorator's
print(double_this.__doc__) # we would expect original function docstrings but.
```

### Overcoming the problem:

### Python has a built-in module **functools** and we can import wraps function. It's a decorator function that is applied to the wrapper function of a decorator. It updates the wrapper to look like WRAPPED function by copying attribute such as mentioned, name,doc and more.

```python
from functools import wraps
```

#### same code but with wraps now:

```python
def decorate_this(func):
    @wraps(func)
    def wrapper(*args): # the wrapper function wraps the original function that we are calling
        """ docstrig wrom the wrapper"""
        print("Starting....")
        result = func(*args)
        print(result)
        print("End.")
        return result

    return wrapper

@decorate_this
def double_this(x):
    """My main function docs"""
    return x**2
d = double_this(4)
```

```python
print(double_this.__name__) # we expect original function name but we will get decorator's
print(double_this.__doc__) # we would expect original function docstrings but.
```

This is useful way of perserving the function originality.
