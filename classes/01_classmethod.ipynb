#### In object-oriented programming, **classes** are objects that have some functionalities.

#### Class have methods and attributes. Withing each class we can have class attributes and object attributes. Methods are functions. 

#### When I construct the class, i use built-in method that is constructor of class. It specifies attributes of an object.

- object is an instance of the class

- class is a blueprint for object creation

```python
# Constructing a simple class:

class Human:
    average_height = 183 # here this variable is CLASS attribute
    def __init__(self,name,age,interest):
        self.name = name # this way we save the attribute in the constructor
        self.age = age # it allows us to call these attributes inside of all different methods within the class
        self.interest = interest
    def __str__(self):
        return f"My name is {self.name} and I am interested in {self.interest}"

    def introduce(self,average_height):
        self.height = average_height
        print(f"My name is {self.name} and I am interested in {self.interest};\nMy class attribute is {self.average_height}")
```

#### __height__: Class attribute, it's explicitly defined inside the class and will stay intact regardless of object. When making an object with this class, constructor won't ask us for this argument.

```python
luka = Human('Luka',27,'programming')
print(luka)
```

### **def __str__ (self)**: string method that is glued to the object. When we print the object, this string returns. It's not typical method such as defined function that does something

```python
luka.introduce()
```

### **But what if we are assigned to use this object within the population of some other country where average height is 170?** 

### How do I change this attribute then?

### **@classmethod** (cls,new_argument)

- cls is the current class

- new_argument is for example new value provided for the attribute to be changed

```python
# @classmethod
# method applied inside the class, with user defined function, which can change the internal value of Class attribute
# Lets check: (copying the same code as above)

class Human:
    average_height = 183 # here this variable is CLASS attribute
    def __init__(self,name,age,interest):
        self.name = name # this way we save the attribute in the constructor
        self.age = age # it allows us to call these attributes inside of all different methods within the class
        self.interest = interest
    def __str__(self):
        return f"My name is {self.name} and I am interested in {self.interest}"

    def introduce(self,average_height):
        self.height = average_height
        print(f"My name is {self.name} and I am interested in {self.interest};\nMy class attribute is {self.average_height}")

    @classmethod
    def change_height(cls,new_height):
        cls.average_height = new_height
        print(cls.average_height)

print(luka.average_height)
```

```python
# Now lets change the height directly inside the class with calling change_height:

Human.change_height(190)
```

```python
Human.average_height
```
