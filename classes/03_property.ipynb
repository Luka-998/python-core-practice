## **Property**

### Python property function allows us to turn class attributes into **properties**

```python
class Person:
    def __init__(self,first_name,last_name):
        self.first_name = first_name
        self.last_name = last_name
        self.email = f"{self.first_name}.{self.last_name}123@gmail.com"
```

```python
obj = Person("Luka","Jasovic")
obj.email
```

```python
obj.first_name = "bob"
obj.email
```

### As we can see here, the email attribute has not been changed. It is because the email is inside the constructor of the class __init__() method and further modification to first_name or last_name will not affect the email attribute

### Email is property of the class __Person__
