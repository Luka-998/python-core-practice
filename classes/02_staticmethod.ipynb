### Class related decorators

## **@staticmethod**

#### @staticmethod do not take **cls** or **self** parameter, although they can take any other arbitary parameter. For example here the function decorated by @staticmethod will take class attribute as the input parameter in order to check the age of the object

#### They **can't** modify class or object state! 

#### Advantage of the @staticmethod is that they do not require object instantiation before a call

```python
class Human:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def hello(self):
        print(f"hello {self.name}")
    @staticmethod
    def is_older(age):
        if age >= 19:
            print("Fine")
```

```python
luka = Human('Luka',49)
luka.is_older(30)
```

### No object created prior calling the class. Simply check the age of the class object.
